## ----------------- Installing Tensorflow Object Detection API -----------------
This page has the official release info https://github.com/tensorflow/models/tree/master/research/object_detection

To properly use this API we need to follow the installation instructions on https://github.com/tensorflow/models/blob/master/research/object_detection/g3doc/installation.md

You can also follow instructions in this video
https://www.youtube.com/watch?v=COlbP62-B-U, or this link https://pythonprogramming.net/introduction-use-tensorflow-object-detection-api-tutorial/

### Install dependencies 

Basically we have TensorFlow and all the rest of the dependencies already, what is missing is the following:
- lxml
- image
So open a CMD and type:
```markdown
pip install lxml
pip install image
```

### Download Protobuf and models

Regarding protobuf go to this page and get the corresponding version, link https://github.com/google/protobuf/releases

Regarding models go to this page and download the reposotory, link https://github.com/tensorflow/models

### Protobuf Compilation
Navigate to directory tensorflow/models/research/ and run the following from a CMD:
```markdown
protoc object_detection/protos/*.proto --python_out=.
```
It didn't work ? are you getting errors? well sometimes it happens, basically here the CMD is not capabale of using the wild character * for some reason. A work around is to compile the proto files one by one... However if you are bored of doing that and want to automate the process then i wrote some python code for that. The code is called 
```markdown
Compiling_all_proto_files.ipynb
```
Run it, ofcourse just edit the **files_path** and **cwd_**
```python
files_path = 'C:/Users/sgs416/Desktop/DL/models/research/object_detection/protos/'
cwd_= "C:/Users/sgs416/Desktop/DL/models/research/"
```

### Running the setup files
Although it is not mentioned, navigate to directory tensorflow/models/research/ open a CMD and type:
```markdown
python setup.py build
python setup.py install
```

people with issues for not doing this step see link https://github.com/tensorflow/models/issues/2031
### Add Libraries to PYTHONPATH
Although it is not mentioned, you still need to add the PYTHONPATH on windows, so do it by following thi link for instructions, link https://stackoverflow.com/questions/3701646/how-to-add-to-the-pythonpath-in-windows-7

Basically this is what you need to do, go to:

```markdown
My Computer > Properties > Advanced System Settings > Environment Variables >
```
Then under system variables create a new Variable called **PythonPath**. In this variable include the following:
```markdown
C:\Users\XXX\Desktop\DL\models;C:\Users\XXX\Desktop\DL\models\research;C:\Users\XXX\Desktop\DL\models\research\slim
```
Obviously you need to point to where the **models**, **research** and **slim** folders are located. These are located in the downloaded the reposotory of https://github.com/tensorflow/models

**Note:** you might need to restart the PC for this to take effect

people with issues for not doing this step see link https://github.com/tensorflow/models/issues/2031
